```python
# -------- GENETIC ALGORITHM CODE (GENERAL TEMPLATE) --------
import numpy as np
import random

# Step 1: Define the objective (fitness) function
# You can change this function as per your problem
def fitness_function(x):
    # Example: maximize f(x) = x * sin(10*pi*x) + 1
    return x * np.sin(10 * np.pi * x) + 1

# Step 2: Initialize parameters
pop_size = 10        # Number of individuals in population
generations = 50     # Number of generations
mutation_rate = 0.1  # Probability of mutation
crossover_rate = 0.8 # Probability of crossover
x_bounds = [0, 1]    # Range of variable x

# Step 3: Initialize population (random values between 0 and 1)
population = np.random.uniform(low=x_bounds[0], high=x_bounds[1], size=pop_size)

# Step 4: Evolution Process
for gen in range(generations):
    # Calculate fitness for each individual
    fitness = fitness_function(population)

    # Step 4.1: Selection (Roulette Wheel)
    probs = fitness / fitness.sum()
    parents = np.random.choice(population, size=pop_size, p=probs)

    # Step 4.2: Crossover
    offspring = []
    for i in range(0, pop_size, 2):
        parent1 = parents[i]
        parent2 = parents[i+1]
        if random.random() < crossover_rate:
            alpha = random.random()
            child1 = alpha * parent1 + (1 - alpha) * parent2
            child2 = alpha * parent2 + (1 - alpha) * parent1
        else:
            child1, child2 = parent1, parent2
        offspring.append(child1)
        offspring.append(child2)
    offspring = np.array(offspring)

    # Step 4.3: Mutation
    for i in range(pop_size):
        if random.random() < mutation_rate:
            offspring[i] += np.random.uniform(-0.1, 0.1)
            offspring[i] = np.clip(offspring[i], x_bounds[0], x_bounds[1])

    # Update population for next generation
    population = offspring

# Step 5: Final result
best_solution = population[np.argmax(fitness_function(population))]
best_fitness = np.max(fitness_function(population))

print("Best solution found:", round(best_solution, 4))
print("Best fitness value:", round(best_fitness, 4))
```
