```python
import numpy as np

A = np.array([0.2, 0.5, 0.7, 1.0])
B = np.array([0.3, 0.7, 0.5, 0.8])

# Basic Operations
union = np.fmax(A, B)
intersection = np.fmin(A, B)
complement_A = 1 - A
complement_B = 1 - B

# Advanced Operations
algebraic_sum = A + B - (A * B)
algebraic_product = A * B
bounded_sum = np.minimum(1, A + B)
bounded_difference = np.maximum(0, A + B - 1)

# Display results
print("A:", A)
print("B:", B)
print("\nUnion:", union)
print("Intersection:", intersection)
print("Complement of A:", complement_A)
print("Algebraic Sum:", algebraic_sum)
print("Algebraic Product:", algebraic_product)
print("Bounded Sum:", bounded_sum)
print("Bounded Difference:", bounded_difference)
```
